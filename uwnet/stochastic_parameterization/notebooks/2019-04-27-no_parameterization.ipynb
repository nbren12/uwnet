```python
import xarray as xr
from uwnet.stochastic_parameterization.compare_stochastic_model_to_base_model import (
    plot_total_pw_over_time,
    plot_equator_pw_over_time,
    plot_npnn_over_time_equator,
    plot_pw_tropics_zonal_variance_over_time,
)
from uwnet.thermo import lhf_to_evap
from uwnet.stochastic_parameterization.utils import get_dataset
dir_ = '/Users/stewart/Desktop/'
ds_s = xr.open_dataset(dir_ + 'stochastic_model_gcm_output.nc')
ds_b = xr.open_dataset(dir_ + 'base_model_gcm_output.nc')
ds_true = get_dataset(
        ds_location="/Users/stewart/projects/uwnet/uwnet/stochastic_parameterization/training.nc",  # noqa
        set_eta=False,
        t_start=0,
        t_stop=len(ds_s.time))
ds_true['NPNN'] = ds_true.Prec - lhf_to_evap(ds_true.LHF)
ds_no_parameterization = xr.open_dataset(dir_ + 'no_parameterization.nc')
%matplotlib inline 
plot_total_pw_over_time(ds_s, ds_b, ds_true, ds_no_parameterization)
plot_equator_pw_over_time(ds_s, ds_b, ds_true, ds_no_parameterization)
plot_npnn_over_time_equator(ds_s, ds_b, ds_true)
plot_pw_tropics_zonal_variance_over_time(ds_s, ds_b, ds_true, ds_no_parameterization)
```
