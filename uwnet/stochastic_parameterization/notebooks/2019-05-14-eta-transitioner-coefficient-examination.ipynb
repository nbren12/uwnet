```python
from uwnet.stochastic_parameterization.stochastic_state_model import EtaTransitioner
dir_ = '/Users/stewart/projects/uwnet/uwnet/stochastic_parameterization/'
base_model_location = dir_ + 'full_model/1.pkl'
ds_location = dir_ + 'training.nc'
transitioner = EtaTransitioner(
    t_start=0,
    t_stop=100,
    poly_degree=1,
    ds_location=ds_location,
    base_model_location=base_model_location
)
transitioner.train()
```

```python
transitioner.transitioner_model[0].coef_.shape
```

```python
%matplotlib inline
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
for idx, predictor in enumerate(transitioner.predictors):
    coefficients = np.array([transitioner.transitioner_model[eta].coef_[:, idx] for eta in transitioner.etas])
    sns.heatmap(
        coefficients,
        cmap='coolwarm',
        xticklabels=transitioner.etas,
        yticklabels=transitioner.etas,
    )
    plt.xlabel('Transitioner Model by Eta')
    plt.ylabel('Coefficient for Transitioning to Bin')
    plt.title('Coefficients for ' + predictor)
    plt.show()
```
