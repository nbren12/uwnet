```python
%matplotlib inline
import matplotlib.pyplot as plt

import src.data


def plot_run_pw_bias(id, field):
    run = src.data.runs[id]

    ng = src.data.open_ngaqua()

    bias = ( run.data_2d[field].mean('x')- ng.data_2d[field].mean('x'))
    rate = (bias[-1] - bias[0])/(bias.time[-1] - bias.time[0])

    rate.plot()
    plt.ylabel("mm/day")
```

```python
plot_run_pw_bias('debias', 'PW')
plot_run_pw_bias('no_debias', 'PW')
plt.legend(['Debiased', 'Not'])
```

```python
plot_run_pw_bias('debias', 'NPNN')
plot_run_pw_bias('no_debias', 'NPNN')
plt.legend(['Debiased', 'Not'])
```
